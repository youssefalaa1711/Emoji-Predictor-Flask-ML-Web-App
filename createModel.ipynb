```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree


emoji_data = pd.read_csv("refined_emojiData.csv")
emoji_data
```

```python
#Split to train model
X= emoji_data.drop(columns="emoji")
y=emoji_data['emoji']

#train data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model =DecisionTreeClassifier()
model.fit(X_train,y_train)

prediction = model.predict([[100, 1]])
prediction
```

```python
prediction1 = model.predict(X_test)

score =accuracy_score(y_test,prediction1)
score
```

```python
joblib.dump(model, "emoji_predictor.joblib")
```
